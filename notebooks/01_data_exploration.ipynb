```python
# Import necessary libraries
import pandas as pd
import numpy as np

# Load the raw dataset
df = pd.read_csv('../data/raw/RoomOccupancyPredictionDataset.csv')  # ../ to move up one level to 'data/raw'

# Explore the dataset
print("Dataset Shape:", df.shape)
df.info()
df.head()
```

```python
df['timestamp'] = pd.date_range(start='1/1/2023', periods=len(df), freq='h')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.lineplot(x='timestamp', y='Occupancy', data=df)
plt.title('Occupancy Over Time')
plt.xlabel('Timestamp')
plt.ylabel('Occupancy')
plt.show()
```

```python
# Fill missing values (if any)
df = df.ffill() 

# Feature engineering: Create new time-based features
df['hour'] = pd.to_datetime(df['timestamp']).dt.hour
df['day_of_week'] = pd.to_datetime(df['timestamp']).dt.dayofweek

# Save processed data
df.to_csv('../data/processed/processed_data.csv', index=False)
print("Processed data saved to data/processed/processed_data.csv")
```
